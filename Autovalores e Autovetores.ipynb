# Autovalores e Autovetores

Um autovetor ($v$) é um vetor não-nulo que quando multiplicado por uma matriz $A$ resulta em uma matriz $\lambda \cdot v$, ou seja, a matriz A possuí uma conjunto de autovetores ($v$) e autovalores ($\lambda$) associado a ele. 

$$
A \cdot \mathbf{v} = \lambda \cdot \mathbf{v} \newline
$$
Onde $A$ é a minha matriz original. Podemos reorganizar como: 

$$
(A - \lambda \cdot \mathbf{I}) \cdot \mathbf{v} = 0
$$

Encontramos os autovalores resolvendo o determinante abaixo: 

$$
\det(A - \lambda \cdot \mathbf{I}) = 0
$$

Após encontrar os autovalores, substituímos na matriz ($\rho - \lambda \cdot \mathbf{I}$) e encontramos as matrizes de autovetores ($v$) associadas ao seu respectivo autovalor ($\lambda$). 

```python
A = np.array([(-3,4), (-1,2)])

Matrix(A)
```

## Polinômio Característico

O polinômio característico de uma matriz quadrada \( A \) é dado por:

$$
p \ (\lambda) = \det \ (A - \lambda I)
$$

Para uma matriz ( $2 \times 2$ ):

$$
A = 
\begin{bmatrix} 
a & b \\ 
c & d 
\end{bmatrix}
$$

O polinômio característico é:

$$
\begin{pmatrix}
\begin{bmatrix} 
a & b \\ 
c & d 
\end{bmatrix}
-
\begin{bmatrix} 
\lambda & 0 \\ 
0 & \lambda
\end{bmatrix}
\end{pmatrix}
.
\begin{bmatrix} 
x_1 \\ 
x_2 
\end{bmatrix}
=
\begin{bmatrix} 
0 \\ 
0 
\end{bmatrix}
$$

O que resulta em:

$$
\begin{bmatrix} 
a - \lambda & b \\ 
c & d - \lambda
\end{bmatrix}
.
\begin{bmatrix} 
x_1 \\ 
x_2 
\end{bmatrix}
=
\begin{bmatrix} 
0 \\ 
0 
\end{bmatrix}
$$

A solução é feito calculando o determinante da matriz com os $\lambda$:

$$
\det\begin{pmatrix} a - \lambda & b \\ c & d - \lambda \end{pmatrix}
=
\lambda^2 - (a + d) \ . \lambda + (a\ .d - b\ .c)
$$

```python
lamda = sy.symbols('lambda')
polinomio = Matrix(A).charpoly(lamda)
polinomio
```
