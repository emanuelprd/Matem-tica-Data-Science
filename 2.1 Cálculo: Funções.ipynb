## Funções

Uma função surge quando precisamos estabelecer uma relação de associação entre variáveis, por exemplo, a receita (R) de uma empresa é uma função do volume de vendas (Q) e do preço do seu produto (P). Uma função tem sempre um domínio que são todos os valores de "entrada" e uma imagem que são todos os valores de "saída" que depende da forma (regra) de como estas variáveis estão relacionadas. 

Por exemplo:

$$
f(x) = x^2 \\
y = x^2 
$$

```python
# Domínio (entrada)
x = np.linspace(-10,10,101)
x
```

```python
# Imagem (saída)
def funcao(x):
    return x**2

funcao(x)
```

### Representação gráfica da função $f(x) = x ^ 2$

```python
plt.plot(x, funcao(x))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.grid()
```

### Funções Simétricas

Uma função chamada de função par quando $f(-x) = f(x)$ e impar quando $f(-x) = -f(x)$.

A função apresentada anteriormente é par, já um exemplo de uma função impar é a função abaixo:

$f(x) = x^3$

```python
def funcao_impar(x):
    return x**3

plt.plot(x, funcao_impar(x))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.grid()
```

### Funções Crescente e Decrescente

Uma função é crescente no intervalo que $f(x_1) < f(x_2)$, por exemplo, a função $f(x) = x^2$ é crescente no intervalo $[0, \infty]$ e decrescente no intervalo $[\infty, 0]$, ou seja, quando $f(x_1) > f(x_2)$.

## Modelos

Um modelo matemático é uma descrição matemática de uma fenômeno da realidade, ele não é uma descrição exata da realidade, mas uma idealização e simplificação de fenômenos reais que permite obter conclusões significativas. 

### Funções Lineares

São modelos que possuem uma inclinação constante ao longo de toda a curva, um exemplo é a equação da reta: 

$$
y = m \cdot x + b \\
f(x) = m \cdot x + b 
$$

```python
def funcao_linear(x):
    return 2 * x + 0

plt.plot(x, funcao_linear(x))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.grid()
```

### Funções Polinômiais

São modelos que podem ser descritos como: 

$$P(x) = a_n x^n + a_{n-1} x^{n-1} + ... + a_2 x^2 + a_1 x^1 + a_0$$

O grau da função é denotado por *n*. 

Um exemplo é a função polinomial de 2º grau é a função abaixo: 

$$f(x) = x^2 + 5 x + 10 $$

```python
def funcao_polinomial(x):
    return x**2 + 5*x + 20

plt.plot(x, funcao_polinomial(x))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.grid()
```

### Funções Potências

São funções que podem ser representadas por: 

$$f(x) = x ^ a$$

Obs. Já representamos anteriormente uma função potência como foi o caso de $f(x) = x^2$ e $f(x) = x^3$.

### Funções Racionais

São representadas como razao de dois polinômios: 

$$f(x) = \frac{P(x)} {Q(x)}$$

```python
def funcao_racional(x):
    return (x**2 + x)/(x - 4)

plt.plot(x, funcao_racional(x))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.grid()
```

### Funções Algébricas

São funções que podem ser representadas por meio de operações algébricas (adição, subtração, multiplicação etc.), exemplo:

$$f(x) = x \sqrt{x+10}$$

```python
def funcao_algebrica(x):
    return x * (x+10)**(0.5)

plt.plot(x, funcao_algebrica(x))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.grid()
```

### Funções Exponenciais

São funções que podem ser descritas por: 

$$f(x) = a^x$$

```python
def funcao_exp(x):
    return 2**x

plt.plot(x, funcao_exp(x))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.grid()
```

### Funções Logarítmicas

São funções que podem ser descritas por: 

$$f(x) = log_a x$$

```python
def funcao_log(x):
    return np.log(x)

plt.plot(x, funcao_log(x))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.grid()
```

## Transformações de Funções

Podemos a partir de uma função transforma-la em uma outra função a partir de certas operações algébricas. Podemos sumarizar possíveis transformações lineares da seguinte forma: 

### Deslocamentos:

$
y = f(x) + c \\
y = f(x) - c \\
y = f(x + c) \\
y = f(x - c)
$

```python
# Deslocamento pra cima
plt.plot(x, funcao(x))
plt.plot(x, funcao(x) + 50)

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.xlim(-10, 10)
plt.ylim(0, 100)

plt.grid()
```

```python
# Deslocamento para o lado
plt.plot(x, funcao(x))
plt.plot(x, funcao(x+5))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.xlim(-10, 10)
plt.ylim(0, 100)

plt.grid()
```

### Reflexões e Expansões:

$
y = c . f(x) \\
y = \frac{1}{c} . f(x) \\
y = f(c . x) \\
y = f(\frac{x}{c}) \\
y = - f(x) \\
y = f(-x)
$

```python
# Expansão Vertical
plt.plot(x, funcao_impar(x))
plt.plot(x, 2*funcao_impar(x))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.xlim(-10, 10)
plt.ylim(-1000, 1000)

plt.grid()
```

```python
# Compressão vertical
plt.plot(x, funcao_impar(x))
plt.plot(x, 1/2 * funcao_impar(x))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.xlim(-10, 10)
plt.ylim(-1000, 1000)

plt.grid()
```

```python
# Compressão horizontal
plt.plot(x, funcao_impar(x))
plt.plot(x, funcao_impar(2*x))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.xlim(-10, 10)
plt.ylim(-1000, 1000)

plt.grid()
```

```python
# Expansão horizontal
plt.plot(x, funcao_impar(x))
plt.plot(x, funcao_impar(x/2))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.xlim(-10, 10)
plt.ylim(-1000, 1000)

plt.grid()
```

```python
# Reflexão no eixo x
plt.plot(x, funcao_impar(x))
plt.plot(x, -funcao_impar(x))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.xlim(-10, 10)
plt.ylim(-1000, 1000)

plt.grid()
```

```python
# Reflexão no eixo y
plt.plot(x, funcao_impar(x))
plt.plot(x, funcao_impar(-x))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.xlim(-10, 10)
plt.ylim(-1000, 1000)

plt.grid()
```

## Combinação de funções

Podemos combinar duas função ou mais funções criando uma função nova função, um exemplo: 

$$(f + g) (x) = f(x) + g(x)$$

Podemos também escrever uma função em função de outra, criamos uma função composta:

$$(f \circ g)(x) = f(g(x))$$

Por exemplo, digamos que $f(x) = x^2$ e $g(x) = x-3$:

$$(f \circ g)(x) = (g(x))^2 = (x-3)^2$$

$$(g \circ f)(x) = (g(x)-3) = x^2-3$$

```python
def funcao_f(x):
    return x**3

def funcao_g(x):
    return x-3

plt.plot(x, funcao_f(funcao_g(x)))
plt.plot(x, funcao_g(funcao_f(x)))

plt.axhline(0, linestyle='-', linewidth=0.9, c='black')
plt.axvline(0, linestyle='-', linewidth=0.9, c='black')

plt.xlabel('X')
plt.ylabel('Y')

plt.xlim(-10, 10)
plt.ylim(-1000, 1000)

plt.grid()
```
